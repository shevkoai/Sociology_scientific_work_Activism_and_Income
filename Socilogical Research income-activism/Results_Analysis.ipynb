## 1. EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

df = pd.read_csv("Sociological_research_form_responces.csv", sep=",")

df.head()
```

```python
df.isnull().sum()
```

```python
# Обработка пропусков
df['Сколько_часов_в_неделю?'] = df['Сколько_часов_в_неделю?'].fillna(0)
df['Сколько_лет_в_СС?'] = df['Сколько_лет_в_СС?'].fillna('Не состоял')
df.iloc[:, 9] = df.iloc[:, 9].fillna('Не состоял ни в одной')  # организации
df.iloc[:, 12] = df.iloc[:, 12].fillna('Не участвовал')  # навыки

df['Скидка'] = df['Скидка'].fillna('Не применимо')
df['Льготник'] = df['Льготник'].fillna('Не применимо')
df.iloc[:, 15] = df.iloc[:, 15].fillna('Не применимо')  # размер скидки
df['Надбавка к стипендии'] = df['Надбавка к стипендии'].fillna('Не применимо')

#Делаю имена колонок короче для удобства
col = {
    df.columns[1]: 'пол',
    df.columns[6]: 'мат_положение',
    df.columns[7]: 'работа_во_время',
    df.columns[8]: 'Состоял_в_Студ._Организациях?',
    df.columns[10]: 'Сколько_часов_в_неделю?',
    df.columns[11]: 'Сколько_лет_в_СС?',
    df.columns[14]: 'Скидка',
    df.columns[18]: 'Трудовой_статус',
    df.columns[21]: 'Руководитель?',
    df.columns[23]: 'Доход',
    df.columns[5]: 'Форма_обучения',
    df.columns[3]: 'образование',
    df.columns[16]: 'Льготник',
    df.columns[17]: 'Надбавка к стипендии',
}
df = df.rename(columns=col)



income_map = {
    'до 1500 BYN': 1000,
    '1500-2000 BYN': 1750,
    '2000-3000 BYN': 2500,
    '3000-5000 BYN': 4000,
    '5000-7500 BYN': 6250,
    '7500-10000 BYN': 8750,
    'Более 10000 BYN': 13000,  # консервативная оценка, но все же
}
df['Доход_числ'] = df['Доход'].map(income_map)

# Формула активизма
df['Активист'] = (
    (df['Сколько_лет_в_СС?'] != 'Не состоял') |
    ((df['Скидка'].isin(['Да, 1 год', 'Да, 2 года', 'Да, 3 года'])) & (df['Льготник'] == 'Нет')) |
    (df['Надбавка к стипендии'].isin(['Да, 1 год', 'Да, 2 года', 'Да, 3 года']))
).astype(int)

df['Сколько_часов_в_неделю?'] = pd.to_numeric(df['Сколько_часов_в_неделю?'], errors='coerce')

wealth_map = {'Ниже среднего': 1, 'Среднее': 2, 'Выше среднего': 3}
df['мат_полож_числ'] = df['мат_положение'].map(wealth_map)

with pd.option_context('display.max_columns', None):
    display(df.head())
```

На этом этапе работы с данными я обработал все пропуски, упростил наиболее частовызываемые колонки для простоты обращения, перевел часы участия в студорганизациях в нумерические значения, а также создал несколько новых колонок - Активист (да/нет, основываясь на моец формуле), нумерическое мат_положение, а также числовой доход (взато по середине промежутка)

## 4. Корреляционный анализ

```python
from scipy.stats import chi2_contingency, spearmanr, pointbiserialr

def cramers_v(var1, var2):
    ct = pd.crosstab(var1, var2)
    chi2, p, dof, expected = chi2_contingency(ct)
    n = ct.sum().sum()
    min_dim = min(ct.shape) - 1
    v = np.sqrt(chi2 / (n * min_dim)) if min_dim > 0 else 0
    return v, p

v, p = cramers_v(df['Состоял_в_Студ._Организациях?'], df['Доход'])
print(f"Cramér's V = {v:.3f}, p = {p:.3f}")



rho, p = spearmanr(df['мат_полож_числ'], df['Доход_числ'], nan_policy='omit')
print(f"Spearman rho = {rho:.3f}, p = {p:.3f}")


mask = df['Доход_числ'].notna()
r, p = pointbiserialr(df.loc[mask, 'Активист'], df.loc[mask, 'Доход_числ'])
print(f"Point-biserial r = {r:.3f}, p = {p:.3f}")
```

## 5. Тест Манна-Уитни

```python
from scipy.stats import mannwhitneyu

mask = df['Доход_числ'].notna()
activists = df.loc[mask & (df['Активист']==1), 'Доход_числ']
non_activists = df.loc[mask & (df['Активист']==0), 'Доход_числ']

stat, p = mannwhitneyu(activists, non_activists, alternative='two-sided')
print(f"Mann-Whitney U = {stat:.1f}, p = {p:.3f}")
print(f"Activists median: {activists.median()}, Non-activists median: {non_activists.median()}")
```

```python
n1, n2 = len(activists), len(non_activists)
r_rb = 1 - (2 * stat) / (n1 * n2)
print(f"Rank-biserial r = {r_rb:.3f}")
```

## 6. Фишер

```python
from scipy.stats import fisher_exact
# For 2x2 tables (e.g., activism × leadership yes/no)
ct = pd.crosstab(df['Активист'], df['Руководитель?'] != 'Нет')
odds_ratio, p = fisher_exact(ct)
print(f"OR = {odds_ratio:.2f}, p = {p:.3f}")
```

```python
# Among activists: does family wealth predict income?
act_df = df[df['Активист']==1 & df['Доход_числ'].notna()]
print(act_df.groupby('мат_положение')['Доход_числ'].median())

# Among non-activists: same
nact_df = df[(df['Активист']==0) & df['Доход_числ'].notna()]
print(nact_df.groupby('мат_положение')['Доход_числ'].median())
```

## 7. Подгрупповой анализ: парадокс Симпсона и когортные эффекты


```python
mask = df['Доход_числ'].notna()

print("=== Медианный доход по когортам выпуска × активизм ===\n")
cohort_table = df[mask].groupby(
    ['Как давно вы закончили бакалавриат?', 'Активист']
)['Доход_числ'].agg(['median', 'mean', 'count'])
display(cohort_table)
```

ВЫВОД: Парадокс Симпсона — внутри каждой когорты активисты зарабатывают 
больше или столько же, но в общей выборке медианы равны. Причина: 
неактивисты непропорционально представлены в старших когортах (3+ лет), 
которые зарабатывают больше просто за счёт стажа.

## 8. Субъективная оценка полезности активизма и доход

```python
# Ключевой вопрос: «Помогло ли участие в студенческих организациях при трудоустройстве?»
help_col = 'Как вы считаете, помогло ли вам участие в студенческих организациях при трудоустройстве? '

print(help_col)


# Бинарная переменная: помогло (напрямую + косвенно) vs остальные
df['Помогло'] = df[help_col].isin([
    'Да, напрямую (упоминал в резюме, нашел через полученные в студенческой организации связи)',
    'Да, косвенно (навыки)'
]).astype(int)

helped = df[mask & (df['Помогло']==1)]['Доход_числ']
not_helped = df[mask & (df['Помогло']==0)]['Доход_числ']

print(f"\n=== Помогло (да) vs Остальные ===")
print(f"Помогло: median={helped.median():.0f}, mean={helped.mean():.0f}, n={len(helped)}")
print(f"Остальные: median={not_helped.median():.0f}, mean={not_helped.mean():.0f}, n={len(not_helped)}")

stat, p = mannwhitneyu(helped, not_helped, alternative='two-sided')
r_rb = 1 - (2 * stat) / (len(helped) * len(not_helped))
print(f"Mann-Whitney U={stat:.1f}, p={p:.3f}, rank-biserial r={r_rb:.3f}")

# Fisher's exact: помогло × доход выше медианы
df['Высокий_доход'] = (df['Доход_числ'] > df.loc[mask, 'Доход_числ'].median()).astype(int)
ct = pd.crosstab(df.loc[mask, 'Помогло'], df.loc[mask, 'Высокий_доход'])
display(ct)

odds, p_fisher = fisher_exact(ct)
print(f"Fisher's exact: OR={odds:.2f}, p={p_fisher:.3f}")
```

```python
# Детализация: напрямую vs косвенно vs нет (среди активистов)
activists_mask = mask & (df['Состоял_в_Студ._Организациях?'] == 'Да, состоял')

print("=== Среди участников организаций: доход по оценке полезности ===\n")
for val in ['Да, напрямую (упоминал в резюме, нашел через полученные в студенческой организации связи)',
            'Да, косвенно (навыки)', 'Нет']:
    sub = df[activists_mask & (df[help_col] == val)]['Доход_числ']
    if len(sub) > 0:
        label = val[:30] + '...' if len(val) > 30 else val
        print(f"  {label}: median={sub.median():.0f}, mean={sub.mean():.0f}, n={len(sub)}")
```
